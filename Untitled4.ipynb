```python
import pandas as pd
import random
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)

data = pd.DataFrame({'whoAmI': lst})
data.head()
```

```python
one_hot = pd.get_dummies(data['whoAmI'])
result = pd.concat([data, one_hot], axis=1)

result.head()
```
